## Formulae: ##
A confidence interval is calculated by finding
$$(point\  \  \  estimate) \pm multiplier\times SE$$

Thus, the confidence interval for one mean is calculated by finding
$$ \bar x \pm t^* \times\frac{s}{\sqrt{n}}$$

Standard error for $\bar{x}$

$$SE(\bar{x})=\frac{\sigma}{\sqrt{n}}$$

Test statistic (in this case t score)

$$t=\frac{\bar{x}-\mu_0}{\frac{s}{\sqrt{n}}}$$

# Question 3 Using built in R functions.#

The `t.test` function may be used for hypothesis tests of a population mean with the appropriate arguments, it also can be used to construct confidence intervals:
* `x` - variable of interest;  
* `mu` - the value of the mean stated in the null hypothesis
* `alternative` - the alternative hypothesis is either "greater", "less", or "two.sided"
* `conf.level` - confidence level desired

Since we do not always need to use all of the options, it is good practice to include the labels for the arguments in the code.

```r
t.test(ncbirths$mage,conf.level=0.95)
```

```r
t.test(ncbirths$mage,conf.level=0.95,alternative="less",mu=27.5)
```
